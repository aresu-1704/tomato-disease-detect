```python
from ultralytics import YOLO
import multiprocessing

def main():
    model = YOLO("yolo12n.pt")
    model.train(
        data='/kaggle/input/vietnameses-tomato-disease-dataset/Tomato-Disease-Dataset/data.yaml',
        epochs=200,
        imgsz=640,
        batch=64,
        pretrained=False,
        device="cuda",
        verbose=True,
        plots=True,
        
        #TƒÉng c∆∞·ªùng
        degrees=10,
        translate=0.1,
        scale=0.5,
        shear=2.0,
        perspective=0.001,
        flipud=0.2,
        fliplr=0.5,
        mosaic=0.8,
        mixup=0.2,
        copy_paste=0.2,
        hsv_h=0.015,
        hsv_s=0.7,
        hsv_v=0.5
    )

if __name__ == '__main__':
    multiprocessing.freeze_support()
    main()
```

```python
from ultralytics import YOLO

model = YOLO("/kaggle/input/yolov12n/pytorch/models/1/best.pt")
model.to('cuda')

data_yaml = "/kaggle/input/tomato-leaf-diseases-dataset-for-object-detection/Tomato leaf diseases dataset for Object Detection/Dataset with augmentation/data.yaml"

if __name__ == '__main__':
    metrics = model.val(data=data_yaml, split='train', save_json=True)

    # In k·∫øt qu·∫£
    print("\n===== Evaluation Metrics =====")
    print(f"Precision (P):     {metrics.box.p.mean()*100:.4f}")
    print(f"Recall (R):        {metrics.box.r.mean()*100:.4f}")
    print(f"F1-score:          {metrics.box.f1.mean()*100:.4f}")
    print(f"mAP@0.5:           {metrics.box.map50.mean()*100:.4f}")
    print(f"mAP@0.5:0.95:      {metrics.box.map.mean()*100:.4f}")
    print("==============================")
```
